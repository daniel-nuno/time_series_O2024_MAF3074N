```python
from turtle import title
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def Kmeans2D(X, K, Figura):
    def distancia(a,b):
        return np.sqrt(np.sum((a-b)**2,axis=1)) # distancia euclideana
    # Definimos los centroides de forma aleatoria
    mx=X[:,0].min()
    Mx=X[:,0].max()
    my=X[:,1].min()
    My=X[:,1].max()
    kx=round(np.sqrt(K))
    ky=np.ceil(K/kx).astype('int')
    ctr=[
        [np.random.uniform(mx+(Mx-mx)/kx*ix,mx+(Mx-mx)/kx*(ix+1)),
        np.random.uniform(my+(My-my)/ky*iy,my+(My-my)/ky*(iy+1))] for ix in range(kx) for iy in range(ky)
    ]
    ctr=np.array(ctr[:K])

    ctra=np.inf*np.ones((K,2))
    while np.sum(np.abs(ctr-ctra))>1e-4:
        Grupos=[]
        for i in range(K):
            Grupos.append([])
        for x in X:
            d=distancia(ctr,x)
            idx=np.argmin(d)
            Grupos[idx].append(x)
        ctra=ctr.copy()
        for i in range(K):
            ctr[i]=np.mean(Grupos[i],axis=0)
    G=[]
    for x in X:
        d=distancia(ctr,x)
        idx=np.argmin(d)
        G.append(idx)
    
    Imagen = np.ones(Figura.shape)
    color = np.linspace(0,1,K)
    for i in range(len(X)):
        Imagen[X[:,0][i], X[:,1][i]] = color[G[i]]
    plt.imshow(Imagen)
    
    for c in ctr:
        plt.scatter(c[1],c[0],marker = 'x',label=np.round(c))
    plt.legend(loc='center left',bbox_to_anchor = (1.01,0.5))
    
    return G,ctr

def cnvrt_img_2d(image_path, U=1):
    #image_path and image name and extension
    #U puede ser de 0 a 255
    Figura = plt.imread(image_path) #la imagen es una matrix 3d con valores de 0 a 255 cada uno representando RGB
    plt.title('original')
    plt.imshow(Figura)
    plt.show()
    Figura = Figura.mean(axis=2) #axis 2 por que es la media de los 3 ejes
    Figura = Figura < U #los valores mayores a U True/False

    df = pd.DataFrame(columns=['x','y'])
    coor = []
    for ix in range(Figura.shape[0]):
        for iy in range(Figura.shape[1]):
            if Figura[ix,iy]: #coordenadas de los valores verdaderos
                coor.append(np.array([ix,iy]))
    df[['x','y']]=coor
    return df, Figura
```

```python
p = 'C:/Users/nuno/OneDrive - ITESO/Ciencia de Datos/ingenieria_de_caracteristicas/discretizacion_variables/kmeans/'
df, Figura = cnvrt_img_2d(p+'danosorios.jpg')
df['kmeans'],ctr = Kmeans2D(df[['x','y']].values,6, Figura)
ctr
```
