```python
for var in df:
  print(float(var)>0)
```

```python
# Definir las listas resultantes
funcionesP=[
            lambda x:1/x,
            lambda x:1/np.sqrt(x),
            lambda x:np.log10(x),
            lambda x:np.sqrt(x)
]
funcionesN=[
            lambda x:x**2,
            lambda x:x**3,
            lambda x:x**4,
            lambda x:x**5
]
SesgoN=[[],[],[],[]]
SesgoP=[[],[],[],[]]
```

```python
for var in df:
  x=df[var]-df[var].min()+1
  xp=[]
  s=[]
  if float(var)>0:
    sxp=SesgoP
    func=funcionesP
  else:
    sxp=SesgoN
    func=funcionesN
  for f in func:
    xp.append(f(x))
    s.append(xp[-1].skew())
  idx=np.argmin(np.abs(s))
  sxp[idx].append((var,s[idx]))
```
