```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ruta='/content/drive/MyDrive/Colab Notebooks/ITESO/Ingeniería de características/P2022_MCD3735A/Módulo 5 - Transformación de variables numéricas/'
```

```python
df=pd.read_csv(ruta+'num.csv')
H=df.hist(bins=50,figsize=(9,7))
```

```python
# Para la variable '1'
#
# Transformación x^2
x=df['1'] # Datos sin desplazar
xp1=x**2
x.skew(),xp1.skew()
```

```python
x2=df['1']-df['1'].min()+1
xp2=x2**2
xp2.skew()
```

```python
x.hist(bins=50)
xp1.hist(bins=50,alpha=0.5)
```

```python
x2.hist(bins=50)
plt.title('Datos desplazados')
plt.figure()
xp2.hist(bins=50)
plt.title('Datos transformados')
```

```python
# Transformación x^3 (sesgo de 'x'=-2.77)
xp1=x**3
xp2=x2**3
xp1.skew(),xp2.skew()
```

```python
# Para la variable '2'
#
# Transformación x^2
x=df['2']-df['2'].min()+1
xp1=x**2
xp2=x**3
x.skew(),xp1.skew(),xp2.skew()
```

```python
x.hist(bins=50)
plt.title('Datos desplazados')

plt.figure()
xp1.hist(bins=50)
plt.title('$x^2$')

plt.figure()
xp2.hist(bins=50)
t=plt.title('$x^3$')
```
