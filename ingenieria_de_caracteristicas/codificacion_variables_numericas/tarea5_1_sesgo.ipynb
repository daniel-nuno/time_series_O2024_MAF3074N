Cargar los datos de `load_boston` de la librería de `scikit-learn`.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import kurtosis, skew, normaltest
from sklearn.datasets import load_boston
all_variables = load_boston().feature_names
all_variables = list(all_variables)
X, y = load_boston(return_X_y=True)
print(X.shape)
```

Primero hay que ver y evaluar la distribución, sesgo, curtosis y normalidad de cada una de las variables. Especificamente queremos evaluar y comparar con un sesgo 0.

```python
evaluate_variables = ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'PTRATIO', 'B', 'LSTAT']
evaluate_index = [all_variables.index(i) for i in evaluate_variables]
evaluate_index
```

```python
sesgos = skew(X)
```

```python
for i in evaluate_index:
    print('skew', skew(X[:,i]))
    print('kurt', kurtosis(X[:,i]))
    print('pvalue', normaltest(X[:,i]).pvalue)
    plt.hist(X[:,i])
    plt.title(all_variables[i])
    plt.show()
```

Las variables con sesgo positivo a la derecha son:
- CRIM
- ZN
- INDUS
- NOX
- RM
- DIS
- LSTAT

Las variables con sesgo negativo a la izquierda son:
- AGE
- PTRATIO
- B


```python
X = X[:,evaluate_index]
```

```python
funciones = [lambda x,n:np.power(x, 1/n), #raiz de n
             lambda x,n:np.log(x) / np.log(n), #log base n
             lambda x,n:x**n] #potencia n

n = list(range(-3,11))
n.remove(0)
sp = np.zeros((len(evaluate_index),
               len(n),
               len(funciones)
              ))
idx = []
for i in range(sp.shape[0]):
    x = X[:,i] - X[:,i].min() + 1
    for j in range(sp.shape[1]):
        y = n[j]
        for k in range(sp.shape[2]):
            f = funciones[k]
            xp = f(x, y)
            sp[i, j, k] = skew(xp)
            
    sp = np.nan_to_num(sp, nan=np.inf)
    idx.append(np.unravel_index(np.argmin(np.abs(sp[i,:,:]), axis=None), sp[i,:,:].shape))
    if idx[-1][1] == 0:
        funcion_name = 'raíz de'
    elif idx[-1][1] == 1:
        funcion_name = 'log base'
    else:
        funcion_name = 'potencia'
    print('La mejor transformación para ' + evaluate_variables[i] + ' usando la función ' + funcion_name + ' n ' + str(n[idx[-1][0]]) )
    f = funciones[idx[-1][1]]
    y = n[idx[-1][0]]
    xp = f(x, y)
    #plt.hist(x, alpha=0.5)
    plt.hist(xp)
    plt.title(evaluate_variables[i])
    plt.show()
```
