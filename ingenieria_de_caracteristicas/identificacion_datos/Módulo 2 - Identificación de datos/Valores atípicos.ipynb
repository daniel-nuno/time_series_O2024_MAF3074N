```python
from sklearn.datasets import load_iris
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

LI=load_iris()
DF=pd.DataFrame(data=LI.data,columns=LI.feature_names)
```

```python
DF.head()
```

```python
DF.shape
```

```python
q1=DF.quantile(0.25)
q3=DF.quantile(0.75)
iqr=q3-q1
iqr
```

```python
DF.describe()
```

```python
tabla=pd.DataFrame(index=DF.columns)
tabla['L_Inferior']=q1-1.5*iqr
tabla['L_superior']=q3+1.5*iqr
tabla
```
