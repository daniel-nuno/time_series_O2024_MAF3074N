Encuentre la mejor función lineal $ Y = a_{0} + a_{1}X_{1} + a_{2}X_{2} $ que pase por los siguientes datos usando la función minimize de la librería scipy y el criterio del promedio de las diferencias al cuadrado utilizando los datos de load_wine.
Donde $ Y $ es igual a la variable *ash*, $ X_{1} $ es *alcalinity_of_ash*, y $ X_{2} $ es *flavanoids*.

```python
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.optimize as opt

LB=load_wine()
data=LB.data
names=LB.feature_names
df=pd.DataFrame(data=data,columns=names)

x1 = df['alcalinity_of_ash']
x2 = df['flavanoids']
y = df['ash']

def fun(a,x1,x2,y):
  f = a[0] + a[1]*x1 + a[2]*x2
  return np.mean((f-y)**2)
```

```python
a = np.random.rand(3)
sol = opt.minimize(fun,a,args=(x1, x2, y))
sol
```

```python
f = sol.x[0] + sol.x[1]*x1 + sol.x[2]*x2
plt.scatter(x1,y)
plt.scatter(x1,f,c='r')
plt.grid()
```

```python
plt.scatter(x2,y)
plt.scatter(x2,f,c='r')
plt.grid()
```
