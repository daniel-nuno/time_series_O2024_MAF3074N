Obtener un DataFrame de pandas a partir del archivo IFC-Subscriptions-and-Voting-Power-of-Member-Count.xml (ubicado en el repositorio de la materia).

```python
import pandas as pd
import xml.etree.ElementTree as ET
ruta = 'C:/Users/nuno/OneDrive - ITESO/Ciencia de Datos/ingenieria_de_caracteristicas/'
```

```python
archivo = ET.parse(ruta + 'IFC-Subscriptions-and-Voting-Power-of-Member-Count.xml')
raiz = archivo.getroot()
type(raiz)
```

```python
for nodo in raiz:
    for subn in nodo:
        print(subn.tag, subn.attrib)
        for subsubn in subn:
            print(subsubn.tag, subsubn.text)
```

```python
dicc = {}
for nodo in raiz:
    for subn in nodo:
        dicc[subn.tag] = []
        for subsubn in subn:
            dicc[subsubn.tag] = []
            
dicc
```

```python
for nodo in raiz:
    for subn in nodo:
        dicc[subn.tag].append(subn.attrib)
        for subsubn in subn:
            dicc[subsubn.tag].append(subsubn.text)

data_frame = pd.DataFrame(dicc)
data_frame.head()
```
